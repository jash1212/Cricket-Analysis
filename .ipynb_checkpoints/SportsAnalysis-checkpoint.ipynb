```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load ODI player data CSV file (change the file path as needed)
df = pd.read_csv("batsman data odi.csv")

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
print("\nPreview of Data:")
print(df.head())

# Data Cleaning
# Convert necessary columns to numeric, handling errors
numeric_columns = ["Matches", "Played", "Not Outs", "Runs", "Batting Average", "Balls Faced", 
                   "Strike Rate", "Centuries", "Fifties", "Ducks", "Fours", "Sixes"]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Drop rows with missing data in important numeric columns
df_cleaned = df.dropna(subset=numeric_columns)

# Basic Statistics
print("\nDescriptive Statistics for Numeric Columns:")
print(df_cleaned[numeric_columns].describe())

# Top Players by Runs
top_run_scorers = df_cleaned.nlargest(10, "Runs")[["Player Name", "Country", "Runs", "Batting Average", "Strike Rate"]]
print("\nTop 10 Players by Runs:")
print(top_run_scorers)

# Top Players by Batting Average (minimum 20 matches)
top_batting_average = df_cleaned[df_cleaned["Matches"] >= 20].nlargest(10, "Batting Average")[["Player Name", "Country", "Batting Average", "Matches"]]
print("\nTop 10 Players by Batting Average (min 20 matches):")
print(top_batting_average)

# Top Players by Strike Rate
top_strike_rate = df_cleaned.nlargest(10, "Strike Rate")[["Player Name", "Country", "Strike Rate", "Runs"]]
print("\nTop 10 Players by Strike Rate:")
print(top_strike_rate)

# Visualization
# Bar plot for top run scorers
plt.figure(figsize=(10, 6))
sns.barplot(x="Runs", y="Player Name", data=top_run_scorers, palette="Blues_d")
plt.title("Top 10 Players by Runs in ODI")
plt.xlabel("Runs")
plt.ylabel("Player Name")
plt.show()

# Scatter plot to compare Strike Rate vs Runs
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Runs", y="Strike Rate", data=df_cleaned, hue="Country", palette="Set1")
plt.title("Strike Rate vs Runs (ODI Players)")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.show()

# Boxplot for Batting Average
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cleaned["Batting Average"], palette="Set2")
plt.title("Distribution of Batting Averages")
plt.show()
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

# Load CSV file (modify to your actual file path)
df = pd.read_csv('batsman data odi.csv')

# Define a dropdown widget to select analysis type
analysis_type = widgets.Dropdown(
    options=['Top Run Scorers', 'Runs Distribution'],
    value='Top Run Scorers',
    description='Analysis Type:',
    disabled=False,
)

# Function to show player analysis
def show_analysis(analysis_type):
    if analysis_type == 'Top Run Scorers':
        display_top_scorers(df)
    elif analysis_type == 'Runs Distribution':
        plot_runs_distribution(df)

# Function to display top 10 run scorers
def display_top_scorers(df):
    top_run_scorers = df.nlargest(10, 'Runs')[["Player Name", "Country", "Runs", "Batting Average", "Strike Rate"]]
    display(top_run_scorers)

# Function to plot runs distribution
def plot_runs_distribution(df):
    plt.figure(figsize=(10, 6))
    df["Runs"].plot(kind="hist", bins=20, color="skyblue", edgecolor='black')
    plt.title("Runs Distribution")
    plt.xlabel("Runs")
    plt.ylabel("Frequency")
    plt.show()

# Create a button to trigger the analysis
button = widgets.Button(description="Show Analysis")

# Define the action for button click
def on_button_click(b):
    show_analysis(analysis_type.value)

button.on_click(on_button_click)

# Display the dropdown and button in Jupyter Notebook
display(analysis_type, button)
```
